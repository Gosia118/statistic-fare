```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean = 3
```

```python
std = 1
```

```python
data_1 = np.random.normal(mean, std, 100)
data_2 = np.random.normal(mean, std, 10000)
```

```python
mean_data_1 = np.mean(data_1)
mean_data_2 = np.mean(data_2)
print(f"Średnia zbioru 100:    {mean_data_1:.2f}")
print(f"Średnia zbioru 10 000: {mean_data_2:.2f}")
```

```python
std_data_1 = np.std(data_1)
std_data_2 = np.std(data_2)
print(f"Odchylenie standardowe zbioru 100:    {std_data_1:.2f}")
print(f"Odchylenie standardowe zbioru 10 000: {std_data_2:.2f}")
```

```python
median_data_1 = np.median(data_1)
median_data_2 = np.median(data_2)
print(f"Mediana zbioru 100:    {median_data_1:.2f}")
print(f"Mediana zbioru 10 000: {median_data_2:.2f}")
```

```python
mode_data_1 = stats.mode(data_1)
mode_data_2 = stats.mode(data_2)
print(f"Moda zbioru 100 {mode_data_1}")
print(f"Moda zbioru 10 000 {mode_data_2}")
```

```python
print("Brak dominanty (mody) w zbiorze 100 oraz zbiorze 10 000.")
```

```python
q0_data_1 = np.quantile(data_1, 0.0)
q1_data_1 = np.quantile(data_1, 0.25)
q3_data_1 = np.quantile(data_1, 0.75)
q4_data_1 = np.quantile(data_1, 1.0)
print(f'''
Najmniejsza wartość zbioru 100: {q0_data_1:.2f}
Największa wartość zbioru 100:  {q4_data_1:.2f}
Pierwszy kwartyl:               {q1_data_1:.2f}
Trzeci kwartyl:                 {q3_data_1:.2f}
''')
```

```python
q0_data_2 = np.quantile(data_2, 0.0)
q1_data_2 = np.quantile(data_2, 0.25)
q3_data_2 = np.quantile(data_2, 0.75)
q4_data_2 = np.quantile(data_2, 1.0)
print(f'''
Najmniejsza wartość zbioru 10 000: {q0_data_2:.2f}
Największa wartość zbioru 10 000:  {q4_data_2:.2f}
Pierwszy kwartyl:                  {q1_data_2:.2f} 
Trzeci kwartyl:                    {q3_data_2:.2f}
''')
```

```python
sns.histplot(data_1, kde=True, bins=30)
plt.title("Histogram dla 100 zmiennych")
plt.xlabel("zmienne x")
plt.ylabel("liczebność")
plt.axvline(x = mean_data_1, color='red', label = 'Średnia')
plt.axvline(x = median_data_1, color='green', label = 'Mediana')
plt.axvline(x = q0_data_1, color='black', label = 'Min')
plt.axvline(x = q1_data_1, color='black', label = 'Q1')
plt.axvline(x = q3_data_1, color='black', label = 'Q3')
plt.axvline(x = q4_data_1, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
print("Dla zbioru 100 rozkład zmiennych jest asymetryczny, negatywnie skośny i występuję przewaga wysokich wartości. Histogram ma rozkład wielomodalnym (multimodalnym), występuje kilka wyraźnych szczytów, a dane są bardzo zróżnicowane.")
```

```python
sns.histplot(data_2, kde=True, bins=30)
plt.title("Histogram dla 10 000 zmiennych")
plt.xlabel("zmienne x")
plt.ylabel("liczebność")
plt.axvline(x = mean_data_2, color='red', label = 'Średnia')
plt.axvline(x = median_data_2, color='green', label = 'Mediana')
plt.axvline(x = q0_data_2, color='black', label = 'Min')
plt.axvline(x = q1_data_2, color='black', label = 'Q1')
plt.axvline(x = q3_data_2, color='black', label = 'Q3')
plt.axvline(x = q4_data_2, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

```python
print("Dla zbioru 10 000 rozkład zmiennych jest normaly (symetryczny). Dane są bardzo skoncentrowane wokół wartości 3.")
```

```python
range_data_1 = max(data_1) - min(data_1)
range_data_2 = max(data_2) - min(data_2)
print(f"Zakres zbioru 100:   {range_data_1:.2f}")
print(f"Zakres zbioru 10 00: {range_data_2:.2f}")
```

```python
iqr_data_1 = q3_data_1 - q1_data_1
iqr_data_2 = q3_data_2 - q1_data_2
print(f"Rozstęp międzykwartylowy zbioru 100:    {iqr_data_1:.2f}")
print(f"Rozstęp międzykwartylowy zbioru 10 000: {iqr_data_2:.2f}")
```

```python
plt.boxplot(data_1)
plt.title("Wykres pudełkowy dla 100 zmiennych")
plt.show()
```

```python
print("Rozkład dla zbioru 100 zmiennych jest lekko lewoskośny. Wąsy są asymetryczne, a mediana znajduję się bardzo blisko środka pudełka. Wartości w zbiorze 100 zmiennych są bardzo rozproszone. Brak wartości odstających.")
```

```python
plt.boxplot(data_2)
plt.title("Wykres pudełkowy dla 10 000 zmiennych")
plt.show()
```

```python
print("Rozkład dla zbioru 10000 zmiennych jest symetryczny. Wąsy mają podobną długość. Wartości są bardziej skoncentrowane, a pudełko jest niższe niż w zbiorze 100 zmiennych. Na wykresie znajduję się dużo wartości odstających.")
```
